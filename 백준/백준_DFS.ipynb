## 문제 1260. DFS와 BFS (O)

```python
from collections import deque

def dfs(graph, visited, start):
    print(start, end=' ')
    visited[start] = True
    
    if len(graph[start]) > 1:
        graph[start].sort()
        
    for i in graph[start]:
        if not visited[i]:
            dfs(graph, visited, i)

def bfs(graph, visited, start):
    q = deque([start])
    visited[start] = True
    while q:
        now = q.popleft()
        print(now, end=' ')
        
        if len(graph[start]) > 1:
            graph[start].sort()
            
        for i in graph[now]:
            if not visited[i]:
                q.append(i)
                visited[i] = True
    
    
n, m, start = map(int, input().split())
graph = [[] for _ in range(n+1)]
visited = [False] * (n+1)
for _ in range(m):
    a, b = map(int, input().split())
    graph[a].append(b)
    graph[b].append(a)
    
dfs(graph, visited, start)
print()
visited = [False] * (n+1)
bfs(graph, visited, start)
```

## 문제 1987. 알파벳 (x, 시간초과)

```python
# 예시 답안, alpha리스트를 통해 중복검사
dx = [-1, 1, 0, 0]
dy = [0, 0, -1, 1]
r, c = map(int, input().split())
graph = [list(map(lambda x: ord(x) - 65, input().rstrip())) for _ in range(r)]
alpha = [0] * 26 

def dfs(start, cnt):
    global result
    result = max(result, cnt)

    x, y = start
    for i in range(4):
        nx = x + dx[i]
        ny = y + dy[i]
        
        if 0 <= nx < r and 0 <= ny < c and alpha[graph[nx][ny]] == 0:
            alpha[graph[nx][ny]] = 1
            dfs((nx, ny), cnt + 1)
            alpha[graph[nx][ny]] = 0


result = 1
alpha[graph[0][0]] = 1
dfs((0,0), 1)
print(result)
```

## 문제 11724. 연결 요소의 개수(O)

```python
def dfs(graph, node):
    visited[node] = True
    
    for i in graph[node]:
        if not visited[i]:
            dfs(graph, i)

n, m = map(int, input().split())
graph = [[] for _ in range(n+1)]
visited = [False] * (n+1)
for _ in range(m):
    a, b = map(int, input().split())
    graph[a].append(b)
    graph[b].append(a)
    
result = 0
for i in range(1, n+1):
    if not visited[i]:
        dfs(graph, i)
        result += 1
        
print(result)
```

## 문제 2468. 안전 영역
