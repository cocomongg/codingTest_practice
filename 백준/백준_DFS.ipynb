## 문제 1260. DFS와 BFS (O)

```python
from collections import deque

def dfs(graph, visited, start):
    print(start, end=' ')
    visited[start] = True
    
    if len(graph[start]) > 1:
        graph[start].sort()
        
    for i in graph[start]:
        if not visited[i]:
            dfs(graph, visited, i)

def bfs(graph, visited, start):
    q = deque([start])
    visited[start] = True
    while q:
        now = q.popleft()
        print(now, end=' ')
        
        if len(graph[start]) > 1:
            graph[start].sort()
            
        for i in graph[now]:
            if not visited[i]:
                q.append(i)
                visited[i] = True
    
    
n, m, start = map(int, input().split())
graph = [[] for _ in range(n+1)]
visited = [False] * (n+1)
for _ in range(m):
    a, b = map(int, input().split())
    graph[a].append(b)
    graph[b].append(a)
    
dfs(graph, visited, start)
print()
visited = [False] * (n+1)
bfs(graph, visited, start)
```

## 문제 2573. 빙산 (O, pypy3)

```python
# import sys
# sys.setrecursionlimit(10**5)
# input = sys.stdin.readline

#동서남북
dx = [0, 0, 1, -1]
dy = [1, -1, 0, 0]


def melt(graph, iceberg):
    remove_list = []
    
    for pos in iceberg:
        for i in range(4):
            nx = pos[0] + dx[i]
            ny = pos[1] + dy[i]
            if graph[nx][ny] == 0 and pos[2] != 0:
                pos[2] -=1

                if pos[2] == 0:
                    remove_list.append(pos)
                else:
                    graph[pos[0]][pos[1]] = pos[2]
                    
    for i in remove_list:
        if i in iceberg:
            graph[i[0]][i[1]] = 0
            iceberg.remove(i)
            
                    
def dfs(graph, iceberg, start):
    x, y = start
    graph[x][y] = 0
    
    for i in range(4):
        nx = x + dx[i]
        ny = y + dy[i]

        if graph[nx][ny] != 0:
            dfs(graph, iceberg, (nx, ny))
            

n, m = map(int, input().split())

iceberg = []
graph = []
for i in range(n):
    graph.append(list(map(int, input().split())))
    for j in range(m):
        if graph[i][j] != 0 :
            iceberg.append([i, j, graph[i][j]])
            
    
result = 0
while True:
    
    # 1. 녹이기
    melt(graph, iceberg)
    result += 1

    # 2. 녹은 빙산 개수 확인
    if len(iceberg) <= 1:
        print(0)
        break
    
    # 3. dfs를 통해 그룹 확인 
    group = 0
    for pos in iceberg:
        if graph[pos[0]][pos[1]] != 0:
            dfs(graph, iceberg, (pos[0], pos[1]))
            group += 1
            
    # 4.그룹 개수 확인
    if group > 1:
        print(result)
        break
    
    else:
        for a in iceberg:
            graph[a[0]][a[1]] = a[2]
```

```python
## 포인트: dp테이블을 사용하여 이미 지나왔던 경로에 대해 중복으로 탐색하는 것을 방지함으로써 시간 단축
##       dp테이블에는 각 칸에서 시작했을 때의 최대값을 저장해 놓음

# import sys
# input = sys.stdin.readline
# sys.setrecursionlimit(10**6)

# 상하좌우
dx = [-1, 1, 0, 0]
dy = [0, 0, -1, 1]

def dfs(x, y):
    if dp[x][y]: # 갱신되었는지 확인 안되어 있다면 동서남북 탐색 수행
        return dp[x][y]
    
    dp[x][y] = 1
    
    for i in range(4):
        nx = x + dx[i]
        ny = y + dy[i]
        
        if nx < 0 or nx >= n or ny < 0 or ny >= n:
            continue
            
        if graph[nx][ny] > graph[x][y]:
            dp[x][y] = max(dp[x][y], dfs(nx, ny) + 1)
        

    return dp[x][y]


n = int(input())
graph = []
for _ in range(n):
    graph.append(list(map(int, input().split())))
    
dp = [[0] * n for _ in range(n)]

result = 0
for x in range(n):
    for y in range(n):
        result = max(result, dfs(x, y))
        
        
print(result)
```

## 문제 11725번. 트리의 부모 찾기 (O, 다시 풀어보기)

```python
# 포인트: 정점으로 주어져서 당황했다. 일단 양방향으로 설정하고, 방문기록을 통해 부모와 자식관계를 파악했다.
# 좀 더 직관적인 풀이로는 if not visitied[i]: parent[node] = i하고 dfs를 호출하는 식으로
# 가능한 이유는 1부터(최고 부모)시작하기 때문에 순차적으로 내려오고, visited로 방문을 검사하기 때문

def dfs(node, visited):
    visited[node] = True
    
    for i in graph[node]:
        if visited[i]:
            parent[node] = i
        else:
            dfs(i, visited)

n = int(input())
parent = [0] * (n+1)
graph = [[] for _ in range(n+1)]
visited = [False] * (n+1)

for _ in range(n-1):
    a, b = map(int,input().split())
    graph[a].append(b)
    graph[b].append(a)

dfs(1, visited)

for parent_node in parent[2:]:
    print(parent_node)
```
