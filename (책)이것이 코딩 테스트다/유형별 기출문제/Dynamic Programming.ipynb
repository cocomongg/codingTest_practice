## 문제 32. 정수 삼각형

```python
# 점화식: dp[i][j] = array[i][j] + max(dp[i-1][j-1], dp[i-1][j])
# 모든 위치를 기준으로 이전 위치로 가능한 2가지 위치까지의 최적의 합 중에서 더 큰 합을 가지는 경우 선택

n = int(input())
dp = []

for i in range(n):
    dp.append(list(map(int, input().split())))
# n = 5
# dp = [
#     [7],
#     [3, 8],
#     [8, 1, 0],
#     [2, 7, 4, 4],
#     [4, 5, 2, 6, 5]
# ]

for i in range(1, n):
    for j in range(i+1):
        print(i, j, end='    ')
        #왼쪽 위에서 내려오는 경우
        if j == 0:
            up_left = 0
        else:
            up_left = dp[i-1][j-1]
            
        #바로 위에서 내려오는 경우
        if j == i:
            up = 0
        else:
            up = dp[i-1][j]
        print(up_left, up, end='    ')
        #최대 합을 저장
        dp[i][j] = dp[i][j] + max(up_left, up)
        print(dp[i][j], end = '    ')
        print()
        
print(dp)
print(max(dp[n-1]))
```

## 문제 33. 퇴사

```python
#점화식: dp[i] = max(p[i] + dp[t[i] + i], max_value)
# 현재 상담 일자의 이윤(p[i]) + 현재 상담을 마친 일자부터의 최대 이윤(dp[t[i] + i])
# dp[i] = i번째 날부터 마지막 날까지 낼 수 있는 최대 이익

n = int(input())

t = [] # 각 상담을 완료하는 데 걸리는 시간
p = [] # 각 상담을 완료했을 때 받을 수 있는 금액
dp = [0] * (n + 1) # 다이나믹 프로그래밍을 위한 1차원 dp 테이블 초기화
max_value = 0 

for _ in range(n):
    x, y = map(int, input().split())
    t.append(x)
    p.append(y)
    

# 리스트를 뒤에서부터 거꾸로 확인
for i in range(n - 1, -1, -1):
    time = t[i] + i
    
    #상담이 기간 안에 끝나는 경우
    if time <= n:
        #점화식에 맞게, 현재까지의 최고 이익 계산
        dp[i] = max(p[i] + dp[time], max_value)
        max_value = dp[i]
        
    #상담기간을 벗어난 경우
    else:
        dp[i] = max_value
        
print(max_value)
```

## 문제 34. 병사 배치하기

```python
# 점화식: 모든 0 <= j < i에 대하여, D[i] = max(D[i], D[j] + 1) if array[j] < array[i]
# D[i] = array[i]를 마지막 원소로 가지는 부분 수열의 최대 길이
# 가장 긴 증가하는 부분 수열로 알려진 알고리즘과 아이디어가 유사

n = int(input())
array = list(map(int, input().split()))

#순서를 뒤집어 '가장 긴 증가하는 부분 수열' 문제로 변환
array.reverse()

#1차원 DP테이블 초기화
dp = [1] * n

# '가장 긴 증가하는 부분 수열(LIS) 알고리즘 수행
for i in range(1, n):
    for j in range(0, i):
        if array[j] < array[i]:
            dp[i] = max(dp[i], dp[j] + 1)
            
            
#열외시켜야 하는 병사의 최소 수 출력
print(n - max(dp))
```

## 문제 35. 못생긴 수

```python
#못생긴 수에 2, 3, 5를 곱합 수 또한 못생긴 수이다.
n = int(input())

ugly = [0] * n #못생긴 수를 담기 위한 테이블(1차원 DP 테이블)
ugly[0] = 1

# 2배, 3배, 5배를 위한 인덱스
i2 = i3 = i5 = 0

#처음에 곱셈값을 초기화
next2, next3, next5 = 2, 3, 5

# 1부터 n까지의 못생긴 수 찾기
for l in range(1, n):
    #가능한 곱셈 결과 중에서 가장 작은 수를 선택
    ugly[l] = min(next2, next3, next5)
 
    #인덱스에 따라서 곱셈 결과를 증가
    if ugly[l] == next2:
        i2 += 1
        next2 = ugly[i2] * 2

    if ugly[l] == next3:
        i3 += 1
        next3 = ugly[i3] * 3

    if ugly[l] == next5:
        i5 += 1
        next5 = ugly[i5] * 5

print(ugly[n-1])
```
