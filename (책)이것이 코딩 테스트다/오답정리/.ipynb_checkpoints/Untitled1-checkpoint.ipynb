```python
from collections import deque

def two_same_char(a, b):
    count = 0
    for i in range(len(a)):
        for j in range(len(b)):
            if a[i] == b[j]:
                count += 1
    if count == 2:
        return True
    return False

def solution(begin, target, words):
    answer = 0
    if target not in words:
        return answer
    
    
    visited = [False] * (len(words)+1)
    start = (begin, 0)
    q = deque([start])
    
    while q:
        now, cost = q.popleft()
        print(now, cost)
        
        if now == target:
            answer = cost
            break
            
        for i in range(len(words)):
            if not visited[i]:
                if two_same_char(now, words[i]):
                    cost += 1
                    q.append((words[i], cost))
                    visited[i] = True
        

    return answer


solution("hit", "cog",["hot", "dot", "dog", "lot", "log", "cog"] )
```

```python
from collections import deque

def two_same_char(a, b):
    count = 0
    for i in range(len(a)):
        for j in range(len(b)):
            if a[i] == b[j]:
                count += 1
    if count == 2:
        return True
    return False

def dfs(visited, graph, v, result, target):
    visited[v] = True
    print(graph[v])
    if graph[v] == target:
        return result
    
    for i in range(len(graph)):
        if not visited[i] and two_same_char(graph[v], graph[i]):
            result += 1
            return dfs(visited, graph, i, result, target)
    return 0
    
def solution(begin, target, words):
    answer = 0
    if target not in words:
        return answer
    
    visited = [False] * (len(words))
    temp = []
    
    now = begin
    while True:
        for i in range(len(words)):
            if not visited[i] and two_same_char(now, words[i]):
                temp.append(words[i])
                visited[i] = True
        if len(temp) == 1:
            answer += 1
            now = temp.pop()
            
        if len(temp) > 1:
            break
    
    save = []
    for i in range(len(temp)):
        result = 0
        result = dfs(visited, words, words.index(temp[i]), result, target)
        if result != 0:
            answer += result
            save.append(answer)

    answer = min(save)
    return answer

solution("hit", "cog",["hot", "dot", "dog", "lot", "log", "cog"] )
```

```python
import heapq

def solution(tickets):
    answer = []
    airport = []
    for i in range(len(tickets)):
        for j in range(2):
            airport.append(tickets[i][j])
    
    airport = list(set(airport))        
    ap_len = len(airport)
    
    graph = {}
    for i in range(len(airport)):
        graph[airport[i]] = []
        
    for i in range(len(tickets)):
        graph[tickets[i][0]].append(tickets[i][1])

    
    q = ['ICN']
    while q:
        now = heapq.heappop(q)
        answer.append(now)
        
        graph[now].sort()
        for i in range(len(graph[now])):
            if graph[now][i] == '':
                continue
            
            heapq.heappush(q, graph[now][i])
            graph[now][i] = ''
            
    return answer
solution([['ICN','AAA'],['ICN','AAA'],['ICN','AAA'],['AAA','ICN'],['AAA','ICN']])
```

```python
def dfs(routes, visited, start, answer, check):
    if len(answer) == check:
        return
    if start not in routes: ## 키가 없는 경우, 경로가 잘못 설정된 경우
        if len(answer) == check - 1:
            answer.append(start)
            return
        else:
            return False
    
    answer.append(start)
    
    for i in range(len(routes[start])):
        if not visited[start][i]:
            visited[start][i] = True
            if not dfs(routes, visited, routes[start][i], answer, check):
                visited[start][i] = False
                
            
def solution(tickets):
    answer = []
    
    routes = dict()
    visited = dict()
    for (start, end) in tickets:
        routes[start] = routes.get(start, []) + [end]
        visited[start] = visited.get(start, []) +[False]
        routes[start].sort()
    start = 'ICN'
    check = len(tickets) + 1
    dfs(routes, visited, start, answer, check)
    return answer
# solution([["ICN", "JFK"], ["HND", "IAD"], ["JFK", "HND"]])
# solution([["ICN", "SFO"], ["ICN", "ATL"], ["SFO", "ATL"], ["ATL", "ICN"], ["ATL","SFO"]])
# solution([['ICN', 'A'], ['ICN', 'B'], ['B', 'ICN']])
# solution([['ICN', 'A'], ['ICN', 'A'], ['A', 'ICN'], ['A' , 'C']])
solution([["ICN", "COO"], ["ICN", "BOO"], ["COO", "ICN"], ["BOO", "DOO"]])
# 입력 : [[ICN, A], [ICN, B], [B, ICN]]
# 답 : [ICN, B, ICN, A]

# 입력 : [[ICN, A], [ICN, A], [A, ICN] [A , C]]
# 답 : [ICN, A, ICN, A, C]
```
